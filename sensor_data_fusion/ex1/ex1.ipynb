```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
v = 20; a_x = 10; a_y = a_z = 1
t = np.linspace(0, a_x/v, 100)

x = v*t
y = a_y*np.sin( (4*math.pi*v*t)/(a_x))
z = a_z*np.sin( (math.pi*v*t)/(a_x))

axes = plt.axes(projection='3d')
axes.plot(x, y, z, label='trajectory', color='b')
axes.legend()
plt.show()
```

Velocity

```python
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d

class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)
```

```python
x_dot = t
y_dot = (4*math.pi*v)/(a_x) * a_y*np.cos( (4*math.pi*v*t)/(a_x))
z_dot = (math.pi*v)/(a_x) * a_z*np.cos( (math.pi*v*t)/(a_x))

axes2 = plt.axes(projection='3d')
for i in range(len(t)-1):
    #axes2.plot(x[i], y[i], z[i], label='', color='b')
    vel_start = [x[i], y[i], z[i]]
    vel_end = [x[i] + x_dot[i], y[i] + y_dot[i], z[i] + z_dot[i]]
    vel_vecs = list(zip(vel_start, vel_end))
    vel_arrow = Arrow3D(vel_vecs[0],vel_vecs[1],vel_vecs[2], mutation_scale=20, lw=1, arrowstyle="-|>", color="y")
    axes2.add_artist(vel_arrow)
axes2.plot(x, y, z)
axes2.legend()
plt.show()
```

```python
x_dot_dot = t
y_dot_dot = (-4*math.pi*v)/(a_x) * (4*math.pi*v)/(a_x) * a_y*np.sin( (4*math.pi*v*t)/(a_x))
z_dot_dot = (-math.pi*v)/(a_x) * (math.pi*v)/(a_x) * a_z*np.sin( (math.pi*v*t)/(a_x))

axes3 = plt.axes(projection='3d')
for i in range(len(t)-1):
    #axes2.plot(x[i], y[i], z[i], label='', color='b')
    vel_start = [x[i], y[i], z[i]]
    vel_end = [x[i] + x_dot_dot[i], y[i] + y_dot_dot[i], z[i] + z_dot_dot[i]]
    vel_vecs = list(zip(vel_start, vel_end))
    vel_arrow = Arrow3D(vel_vecs[0],vel_vecs[1],vel_vecs[2], mutation_scale=20, lw=1, arrowstyle="-|>", color="g")
    axes3.add_artist(vel_arrow)
axes3.plot(x, y, z)
axes3.legend()
plt.show()
```
